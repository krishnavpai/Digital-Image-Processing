```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('images/building.tif', cv2.IMREAD_GRAYSCALE)

# board = cv2.cvtColor(board, cv2.COLOR_BGR2RGB)
plt.imshow(image, cmap = 'gray')
plt.axis('off')
```

```python
# Function to apply edge operators
def apply_edge_operators(image):

    # Prewitt operators
    prewitt_kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    prewitt_kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
    prewitt_Gx = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_x)
    prewitt_Gy = cv2.filter2D(image, cv2.CV_64F, prewitt_kernel_y)

    # Sobel operators

    # sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    # sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    sobel_kernel_x = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
    sobel_kernel_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
    sobel_Gx = cv2.filter2D(image, cv2.CV_64F, sobel_kernel_x)
    sobel_Gy = cv2.filter2D(image, cv2.CV_64F, sobel_kernel_y)



    # Roberts operators
    roberts_kernel_x = np.array([[1, 0], [0, -1]])
    roberts_kernel_y = np.array([[0, 1], [-1, 0]])
    roberts_Gx = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_x)
    roberts_Gy = cv2.filter2D(image, cv2.CV_64F, roberts_kernel_y)

    return sobel_Gx, sobel_Gy, prewitt_Gx, prewitt_Gy, roberts_Gx, roberts_Gy

```

```python
# Apply edge operators
sobel_Gx, sobel_Gy, prewitt_Gx, prewitt_Gy, roberts_Gx, roberts_Gy = apply_edge_operators(image)

```

```python
# Canny Edge Detection
edges = cv2.Canny(image, 100, 200)  # You can adjust the thresholds as needed
```

```python
# Display the original and processed images
plt.figure(figsize=(10, 5))

plt.subplot(1,2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap = 'gray')
plt.title('Segmented Image (Edges) using canny')
plt.axis('off')

```

```python
plt.figure(figsize=(21, 5))

plt.subplot(1, 4, 1)
plt.imshow(sobel_Gx, cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(sobel_Gy, cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(np.sqrt(sobel_Gx**2 + sobel_Gy**2), cmap='gray')
plt.title('Sobel Magnitude')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')
```

```python
plt.figure(figsize=(21, 5))

plt.subplot(1, 4, 1)
plt.imshow(prewitt_Gx, cmap='gray')
plt.title('Prewitt X')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(prewitt_Gy, cmap='gray')
plt.title('Prewitt Y')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(np.sqrt(prewitt_Gx**2 + prewitt_Gy**2), cmap='gray')
plt.title('Prewitt Magnitude')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')
```

```python
plt.figure(figsize=(21, 5))

plt.subplot(1, 4, 1)
plt.imshow(roberts_Gx, cmap='gray')
plt.title('Roberts X')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(roberts_Gy, cmap='gray')
plt.title('Roberts Y')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(np.sqrt(roberts_Gx**2 + roberts_Gy**2), cmap='gray')
plt.title('Roberts Magnitude')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(image, cmap='gray')
plt.title('Original')
plt.axis('off')
```

```python
finger = cv2.imread('images/noisy_fingerprint.tif', cv2.IMREAD_GRAYSCALE)

# Apply GaussianBlur to reduce noise and improve Otsu's thresholding
blurred = cv2.GaussianBlur(finger, (5, 5), 0)

# Use Otsu's thresholding
_, thresholded = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the original and thresholded images
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(finger, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(blurred, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(thresholded, cmap='gray')
plt.title('Otsu Thresholding')
plt.axis('off')

plt.show()
```
