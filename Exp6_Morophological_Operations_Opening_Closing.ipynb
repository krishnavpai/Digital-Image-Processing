```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
board = cv2.imread('images/wire_board.tif', cv2.IMREAD_GRAYSCALE)

# board = cv2.cvtColor(board, cv2.COLOR_BGR2RGB)
plt.imshow(board, cmap = 'gray')
```

```python
# Define a kernel for erosion and dilation
kernel_3 = np.ones((3, 3), np.uint8)

kernel_7 = np.ones((7, 7), np.uint8)
```

```python
# Perform erosion
erosion_3 = cv2.erode(board, kernel_3, iterations=1)

# Perform dilation
dilation_3 = cv2.dilate(board, kernel_3, iterations=1)

# Perform erosion
erosion_7 = cv2.erode(board, kernel_7, iterations=1)

# Perform dilation
dilation_7 = cv2.dilate(board, kernel_7, iterations=1)

```

```python
# Display the original, erosion, and dilation images
fig = plt.figure(figsize=(15, 7))
plt.subplot(1, 3, 1)
plt.title('Original', fontsize = '11')
plt.axis('off')
plt.imshow(board, cmap = 'gray')

plt.subplot(2, 3, 2)
plt.title('Eroded (3x3)', fontsize = '11')
plt.axis('off')
plt.imshow(erosion_3, cmap = 'gray')

plt.subplot(2, 3, 3)
plt.title('Dilated (3x3)', fontsize = '11')
plt.axis('off')
plt.imshow(dilation_3, cmap = 'gray')

plt.subplot(2, 3, 5)
plt.title('Dilated (7x7)', fontsize = '11')
plt.axis('off')
plt.imshow(erosion_7, cmap = 'gray')

plt.subplot(2, 3, 6)
plt.title('Dilated (7x7)', fontsize = '11')
plt.axis('off')
plt.imshow(dilation_7, cmap = 'gray')

```

```python
fingerprint = cv2.imread('images/noisy_fingerprint.tif', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(fingerprint, cmap = 'gray')
```

```python
#opening
# Perform erosion
step1 = cv2.erode(fingerprint, kernel_3, iterations=1)

# Perform dilation
step2 = cv2.dilate(fingerprint, kernel_3, iterations=1)
```

```python
opening = step2
```

```python
#closing
# Perform erosion
step3 = cv2.dilate(fingerprint, kernel_3, iterations=1)

# Perform dilation
step4 = cv2.erode(fingerprint, kernel_3, iterations=1)
```

```python
closing = step4
```

```python
plt.figure(figsize = (15, 9))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(fingerprint, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Opening")
plt.imshow(opening, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Closing")
plt.imshow(closing, cmap='gray')
plt.axis('off')
```
